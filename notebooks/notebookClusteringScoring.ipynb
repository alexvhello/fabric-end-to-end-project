```python
from pyspark.ml import PipelineModel
from pyspark.ml.clustering import KMeansModel
from pyspark.sql.functions import col
```

```python
# Cargar pipeline y modelo de clustering

pipeline_loaded = PipelineModel.load("Files/Models/pipeline_preprocesamiento")
kmeans_model = KMeansModel.load("Files/Models/Clustering_KMeans")

print("✅ Pipeline y modelo de clustering cargados.")
```

```python
# Cargar datos nuevos desde Silver

dfNuevos = spark.sql("SELECT * FROM lakehouseSilver.dfcontratosfinal")

print(f"✅ Nuevos contratos cargados. Filas: {dfNuevos.count()}")
```

```python
# Aplicar pipeline para generar scaledFeatures

dfNuevos_preparado = pipeline_loaded.transform(dfNuevos)

print("✅ Nuevos datos transformados (scaledFeatures listos).")
```

```python
# Asignar cluster a cada contrato

predicciones = kmeans_model.transform(dfNuevos_preparado)
```

```python
# Seleccionar columnas finales

resultado = predicciones.select("numContrato", "idCliente", col("prediction").alias("cluster"))
```

```python
# Guardar tabla optimizada para Power BI

resultado.write.format("delta").mode("overwrite").saveAsTable("Predicciones_Clusters")

print("✅ Tabla 'Predicciones_Clusters' lista para Power BI.")
```
