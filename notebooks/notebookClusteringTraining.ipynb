```python
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator
```

```python

# --------------------
# 1. Cargar dataset
# --------------------
dfCluster = spark.table("ML_Contratos_Clustering")
```

```python

# --------------------
# 2. Determinar número óptimo de clusters
# --------------------
silhouette_scores = []
for k in range(2, 8):  # Probamos de 2 a 7 clusters
    kmeans = KMeans(featuresCol="scaledFeatures", k=k, seed=42)
    model = kmeans.fit(dfCluster)
    predictions = model.transform(dfCluster)
    
    evaluator = ClusteringEvaluator(featuresCol="scaledFeatures", metricName="silhouette", distanceMeasure="squaredEuclidean")
    score = evaluator.evaluate(predictions)
    silhouette_scores.append((k, score))

print("✅ Silhouette Scores:")
```

```python
for k, score in silhouette_scores:
    print(f"K={k}, Silhouette={score:.4f}")

# Elegimos el mejor K
best_k = max(silhouette_scores, key=lambda x: x[1])[0]
print(f"✅ Mejor número de clusters: {best_k}")
```

```python

# --------------------
# 3. Entrenar modelo KMeans con best_k
# --------------------
kmeans = KMeans(featuresCol="scaledFeatures", k=best_k, seed=42)
kmeans_model = kmeans.fit(dfCluster)
```

```python


# --------------------
# 4. Guardar modelo en Lakehouse
# --------------------
kmeans_model.write().overwrite().save("Files/Models/Clustering_KMeans")

print("✅ Modelo de clustering guardado en Lakehouse.")
```
